```python
def radix_sort(A, d, k):
    # A consists of n d-digits, with digits ranging from 0 to k-1
    for j in range(d):
        A_j = []
        for item in A:
            if j>=1:
                A_j.append(((int)((item%(10**(j+1)) - item%(10**j))/10**j), item))
            else:
                A_j.append((item%10, item))
        count = [0]*k
        nc = [0]*(k+1)
        for tupla in A_j:
            count[tupla[0]]+=1
        for i in range(1,k):
            nc[ i ] = nc[ i - 1 ] + count[ i - 1 ]
        nc[ k ] = len(A)
        ncfixo = nc.copy()
        nc1 = nc.copy()
        result = A_j.copy()
        for i in range(k):
            while nc[ i ]<ncfixo[ i + 1 ]:
                aux = A_j[ nc[ i ] ][0]
                result[ nc1[ aux ] ] = A_j[ nc[ i ] ]
                nc[ i ]+=1
                nc1[ aux ]+=1
        A=[]
        for tupla in result:
            A.append(tupla[1])
    return A
```

```python
A = [201, 10, 3, 100]
A_sorted = radix_sort(A, 3, 10)
print(A_sorted)
```
